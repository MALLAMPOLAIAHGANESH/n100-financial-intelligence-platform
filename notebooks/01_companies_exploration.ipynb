## 1. Import Libraries

We install and import the necessary Python libraries required for data manipulation (`pandas`, `numpy`) and set standard Pandas display options so that large tables aren't truncated.

```python
# Install all required dependencies for reading and styling
%pip install pandas numpy openpyxl jinja2 matplotlib

import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 200)
```

## 2. Project Path

We dynamically locate the project root directory and define the path to our raw dataset (`companies.xlsx`). This code automatically detects if you are running it from inside the `notebooks` folder or the project root!

```python
current_dir = Path.cwd()

# Dynamically determine project root
if current_dir.name == "notebooks":
    PROJECT_ROOT = current_dir.parent
else:
    PROJECT_ROOT = current_dir

DATA_PATH = PROJECT_ROOT / "data" / "raw"
COMPANY_FILE = DATA_PATH / "companies.xlsx"

print("Project Root:", PROJECT_ROOT)
print("Company File:", COMPANY_FILE)
```

## 3. Read Dataset

We load the Excel dataset into a Pandas DataFrame so we can begin our exploration and manipulation.

```python
try:
    df = pd.read_excel(COMPANY_FILE)
    print("Dataset Loaded Successfully")
    display(df.head())
except Exception as e:
    print(f"Error loading dataset: {e}")
```

## 4. Dataset Overview

We print the fundamental shape of our dataset to understand its scale (total number of rows and columns).

```python
print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

## 5. Column Names

We extract and list all the column names present in the dataset. This acts as our initial data dictionary.

```python
columns = pd.DataFrame({
    "Column Name": df.columns
})
columns
```

## 6. Data Types

We inspect the data types of each column (e.g., `int64`, `float64`, `object`/string) to ensure Pandas interpreted them correctly.

```python
dtype_df = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.astype(str)
})
dtype_df
```

## 7. Missing Values

We calculate the absolute count and percentage of missing (null/NaN) values for each column. High missing percentages might require imputation or dropping the column entirely.

```python
missing = pd.DataFrame({
    "Column": df.columns,
    "Missing Values": df.isnull().sum(),
    "Percentage": round(df.isnull().mean()*100,2)
})

missing.sort_values("Missing Values", ascending=False)
```

## 8. Duplicate Records

We check for completely identical rows across all columns, which could skew our analysis if left untreated.

```python
duplicates = df.duplicated().sum()
print(f"Duplicate Rows : {duplicates}")
```

## 9. Unique Values

We count the number of distinct, unique values in each column. This helps identify categorical variables or unique identifiers (like a Company ID).

```python
unique = pd.DataFrame({
    "Column": df.columns,
    "Unique Values": df.nunique()
})
unique
```

## 10. Descriptive Statistics

We generate summary statistics (mean, median, standard deviation, min, max, and quartiles) to understand the distribution of our numerical columns.

```python
df.describe(include="all").T
```

## 11. Sample Records

We randomly sample a few records to visually inspect the actual contents and context of the data.

```python
df.sample(10, random_state=42)
```

## 12. Complete Data Profile

We consolidate all the column metadata we just explored (data types, missing values, unique counts) into a single profile DataFrame for a unified view.

```python
profile = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.astype(str),
    "Missing": df.isnull().sum(),
    "Missing %": round(df.isnull().mean()*100,2),
    "Unique": df.nunique()
})
profile
```

## 13. Export Profile

We save our consolidated data profile as a CSV report so we can review it later or share it with stakeholders.

```python
REPORT_PATH = PROJECT_ROOT / "reports"
REPORT_PATH.mkdir(exist_ok=True)

profile.to_csv(
    REPORT_PATH / "companies_profile.csv",
    index=False
)

print("Profile Exported Successfully")
```

## 14. Data Quality Checks

We summarize the overall health of the dataset by restating the total rows, total duplicates, and a breakdown of all missing values.

```python
print("=" * 60)
print("DATA QUALITY CHECK")
print("=" * 60)

print("Total Rows :", len(df))
print("Duplicate Rows :", df.duplicated().sum())
print("Total Missing :", df.isnull().sum().sum())

display(df.isnull().sum())
```

## 15. Save Clean Copy

Finally, we export the DataFrame as a `.csv` file to our processed data folder so it is ready for the next stages of financial analysis.

```python
PROCESSED = PROJECT_ROOT / "data" / "processed"
PROCESSED.mkdir(exist_ok=True)

df.to_csv(
    PROCESSED / "companies.csv",
    index=False
)

print("Clean CSV Saved Successfully")
```
