## 1. Import Libraries

We import `pandas` and `pathlib` for file routing and structured data analysis.

```python
import pandas as pd
from pathlib import Path
import sys

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

print("Libraries Loaded Successfully")
```

## 2. Set Project Paths

Establish the absolute paths to locate our raw datasets and supporting spreadsheets, dynamically resolving relative to the Jupyter execution folder.

```python
current_dir = Path.cwd()
PROJECT_ROOT = current_dir.parent if current_dir.name == "notebooks" else current_dir

RAW_DATA = PROJECT_ROOT / "data" / "raw"
SUPPORTING_DATA = PROJECT_ROOT / "data" / "supporting"

print(f"Project Root: {PROJECT_ROOT}")
print(f"Raw Data Path: {RAW_DATA}")
print(f"Supporting Data Path: {SUPPORTING_DATA}")
```

## 3. Discover All Excel Files

Scan both folders for Excel spreadsheets and compile a master list of files to analyze.

```python
raw_files = sorted(RAW_DATA.glob("*.xlsx"))
supporting_files = sorted(SUPPORTING_DATA.glob("*.xlsx"))

excel_files = raw_files + supporting_files

print(f"Total Datasets: {len(excel_files)}")

for file in excel_files:
    print(file.name)
```

## 4. Analyze Every Dataset

Read each Excel file to retrieve structural metadata: row counts, column counts, and column header names.

```python
summary = []

for file in excel_files:
    try:
        df = pd.read_excel(file)
        summary.append({
            "Dataset": file.name,
            "Rows": len(df),
            "Columns": len(df.columns),
            "Column Names": ", ".join(map(str, df.columns))
        })
    except Exception as e:
        print(f"Failed to read {file.name}: {e}")

summary_df = pd.DataFrame(summary)
display(summary_df)
```

## 5. Save Mapping Report

Export the structural analysis to `/reports/dataset_mapping_report.csv` for structural mapping tasks.

```python
REPORT_PATH = PROJECT_ROOT / "reports"
REPORT_PATH.mkdir(exist_ok=True)

summary_df.to_csv(
    REPORT_PATH / "dataset_mapping_report.csv",
    index=False

)

print("Dataset Mapping Report Generated Successfully!")
```

## 6. Inspect Master Dataset in Detail

Analyze the master `companies.xlsx` file. We print its head, raw data types, and check for missing fields.

```python
df_comp = pd.read_excel(RAW_DATA / "companies.xlsx")

display(df_comp.head())

print("\n--- Data Types ---")
print(df_comp.dtypes)

print("\n--- Missing Values ---")
print(df_comp.isnull().sum())
```

## 7. Inspect Financial Datasets in Detail

Print detailed summaries (head, dtypes, nulls) for `profitandloss.xlsx`, `balancesheet.xlsx`, and `cashflow.xlsx`.

```python
financials = ["profitandloss.xlsx", "balancesheet.xlsx", "cashflow.xlsx"]

for fn in financials:
    print("=" * 80)
    print(f"FINANCIAL FILE: {fn}")
    print("=" * 80)
    
    try:
        df_f = pd.read_excel(RAW_DATA / fn)
        display(df_f.head(3))
        print("\nData Types:")
        print(df_f.dtypes)
        print("\nMissing Values:")
        print(df_f.isnull().sum())
        print("\n")
    except Exception as e:
        print(f"Could not load {fn}: {e}\n")
```
