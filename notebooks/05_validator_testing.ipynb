## 1. Imports

We inject the project root into the system path to allow importing our custom `DataValidator` module and standard libraries.

```python
import pandas as pd
from pathlib import Path
import sys

current_dir = Path.cwd()
PROJECT_ROOT = current_dir.parent if current_dir.name == "notebooks" else current_dir

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.etl.validator import DataValidator

print("Libraries Imported Successfully")
```

## 2. Project Paths

Establish absolute paths for our raw data and reporting directories.

```python
DATA = PROJECT_ROOT / "data" / "raw"
REPORT = PROJECT_ROOT / "reports"
REPORT.mkdir(exist_ok=True)

print(f"Raw Data Path: {DATA}")
```

## 3. Load Dataset

Load the master `companies.xlsx` dataset to serve as our primary validation test subject.

```python
file = DATA / "companies.xlsx"
df = pd.read_excel(file)

display(df.head())
```

## 4. Run Validator

Instantiate our advanced `DataValidator` class and execute the validation pipeline across the dataset. This triggers rules for empty sets, duplicate columns/rows, missing values, date ranges, and URL patterns.

```python
validator = DataValidator()

# Execute the validation rules
df = validator.validate(df, "companies")

print("Validation rules applied successfully.")
```

## 5. Review Validation Report

Fetch and display the high-level summary report of all validation checks performed.

```python
report = validator.get_report()
display(report)
```

## 6. Review Failures

Fetch and display specific failure instances (like missing values, duplicate rows, or malformed URLs) caught by our strict rules.

```python
failures = validator.get_failures()
display(failures)
```

## 7. Export Reports

Persist our data quality findings to CSV files inside the `/reports` directory for review by data engineering teams.

```python
if not report.empty:
    report.to_csv(REPORT / "validation_report.csv", index=False)

if not failures.empty:
    failures.to_csv(REPORT / "validation_failures.csv", index=False)

print("Reports Generated and Saved to Disk!")
```
